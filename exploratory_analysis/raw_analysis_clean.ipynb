```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df_no_str=df_clean.drop(['School'], axis=1)
no_na= df_no_str.dropna()
no_na
```

```python
labels = no_na['highest_degree']
labels2 = no_na['ownership']
df2=no_na.drop(['ownership'], axis=1)
df2
```

```python
# # Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df2)
```

```python
tsne = TSNE(learning_rate=35)
```

```python
tsne_features = tsne.fit_transform(X_scaled)
```

```python
# The dataset has 2 columns
tsne_features.shape
```

```python
# Prepare to plot the dataset

# The first column of transformed features
df2['x'] = tsne_features[:,0]

# The second column of transformed features
df2['y'] = tsne_features[:,1]
```

```python
# Visualize the clusters
plt.scatter(df2['x'], df2['y'])
plt.show()
```

```python
labels2.value_counts()
```

```python
# Visualize the clusters with color
plt.scatter(df2['x'], df2['y'], c=labels)
plt.show()
```

```python
pca = PCA(n_components=8)
principal_components = pca.fit_transform(df2)
principal_components.shape
```

```python
# The sum of the explained variance of the principal components
pca.explained_variance_.sum()
```

```python
# Reduce dimensions with t-SNE
tsne = TSNE(perplexity=50)
tsne_features = tsne.fit_transform(principal_components)
tsne_features.shape
```

```python
# Plot t-SNE output
X = tsne_features[:,0]
y = tsne_features[:,1]

plt.scatter(X, y)
plt.show()
```

```python
# Identify clusters with k-means
inertia = []
k = list(range(1,20))

for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(principal_components)
    inertia.append(kmeans.inertia_)
```

```python
# Create an elbow plot
elbow_df = pd.DataFrame({'k': k, 'inertia': inertia})
elbow_df.plot.line(x='k', y='inertia')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

```python
X = df2
y = labels2
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
features = clf.feature_importances_
print(features)
plt.bar(x = range(len(features)), height=features)
plt.show()
```

```python
from sklearn.feature_selection import SelectFromModel
sel = SelectFromModel(clf)
sel.fit(X_train_scaled, y_train)
sel.get_support()
```

```python
X_selected_train, X_selected_test, y_train, y_test = train_test_split(sel.transform(X), y, random_state=1)
scaler = StandardScaler().fit(X_selected_train)
X_selected_train_scaled = scaler.transform(X_selected_train)
X_selected_test_scaled = scaler.transform(X_selected_test)
```

```python
clf = LogisticRegression()
clf.fit(X_selected_train_scaled, y_train)
print(f'Training Score: {clf.score(X_selected_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_selected_test_scaled, y_test)}')
```
