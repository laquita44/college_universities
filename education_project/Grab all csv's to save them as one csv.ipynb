```python
#  Dependencies
import pandas as pd
from glob import glob
```

```python
#  Use glob to get all csv files at one time
edu = sorted(glob('data/data_file_*.csv'))
edu
                        
```

```python
df = pd.concat((pd.read_csv(file).assign(filename = file)
                for file in edu), ignore_index = True)
df.head()
```

```python
bexiting = sorted(glob('data/data_file_*.csv'))
bexiting
```

```python
df1 = pd.concat((pd.read_csv(file).assign(filename = file)
                for file in bexiting), ignore_index = True)
df1.head()
```

```python
cexiting = sorted(glob('data/data_file_*.csv'))
cexiting
```

```python
df2 = pd.concat((pd.read_csv(file).assign(filename = file)
                for file in cexiting), ignore_index = True)
df2.head()
```

```python
# save DataFrames as a csv file
df.to_csv('both_b_c_exiting.csv', index=False)
df1.to_csv('bexiting.csv', index=False)
df2.to_csv('cexiting.csv', index=False)
```
