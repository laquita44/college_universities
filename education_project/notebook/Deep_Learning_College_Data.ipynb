```python
# Drop the non-beneficial ID columns, 'NAME'.
df =df.drop(columns=["Name"])
df.head()
```

```python
# Transform Previous Customer column
def changeStatus(status):
    if status == 'Yes':
        return 1
    else:
        return 0
# Along with replace() and map(), this is another way to encode the gender column into numbers.
df['Private'] = df['Private'].apply(changeStatus)
```

```python
y =df['Private'].values
y
```

```python
X =df.drop('Private', axis=1).values
X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
